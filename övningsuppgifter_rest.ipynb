# Kapitel 1 - Introduktion till maskininlärning


1. Hur hänger AI, ML och DL ihop?

   Ai är programmerad att efterlikna människorseller djurs intelligens.
   ML är en del av AI, där ML gör att datorn kan lära sig utan att vara explicit programmerad.
   Och DL i sin tur är en del av ML, där modeller som kallas för neurala nätverk används. 
   Neurala nätverk har en modellarkitektur som ursprungligen var inspirerade av hur människohjärnan fungerar.

   ___________________________________________________________________________________
   

2. Vilka är de fyra problemkategorierna inom ML? 

   Regression - kontinuerliga värden

   Klassificering - diskreta värden

   Dimensionsreducering - Förenkla data, överföra den til en rymd med lägre dimensionalitet.
   Detta gör man för att modellträningen ska gå snabbare (regressionsmodell eller klassificeringsmodeller)

   Klustering - Endast indata.

3. Förklara följande: 


a) Vad är syftet med att dela upp data i träningsdata, valideringsdata och testdata? 

För att kunna utvärdera våra modeller är det viktigt att vi har tillgång till data som modellen inte har tränats på. Testdatan representerar datan som den färdiga modellen ska använda för att göra prediktioner.

___________________________________________________________________________________

b) Vad är k-delad korsvalidering för något? 

En metod där man delar upp datan in träningsdata och testdata. 80-20 eller 70-30.
Träningsdatan delas därefter upp i k lika stora delar där modellen tränas k gånger. K är vanligtvis 5 eller 10. Att använda k-delad korsvalidering kan ge en bättre skattning på hur bra modellen presterar. Nackdelen är att varje modell behöver tränas k gånger, som kan ta lång tid.

___________________________________________________________________________________

c) Vad är *root mean squared error* (*RMSE*) för något?

RMSE är ett utvärderingsmått för regressionsmodeller och dess matematiska uttryck.

RMSE används för att tolka felet som modeller gör eller för att rangordna modeller - vilken är bättre? Vi kan också använda MSE för att rangordna modeller. Skillnaden är roten-ur tecknet, där kvadratroten utgör en extra operation som kan ta längre tid att genomföra.

___________________________________________________________________________________

d) Vad är en hyperparameter för något? Vad är en parameter för något? 

En hyperparatmeter svarar på frågan "hur vi lärt oss" och en parameter svarar på frågan "vad vi lärt oss".

___________________________________________________________________________________

e) Vad är *grid search* och hur hänger namnet *"grid"* och *"search"* ihop med processen som genomförs? Läser vi dokumentationen: 
[https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV)
så ser vi att hyperparametern `refit` har standardvärdet `True`, vad innebär det? 

Grid search - ett automatiserat sätt att välja hyperparametrar i ett rutnät ("grid") och letar efter den bästa kombinationen ("search").

refit=True betyder att modellen tränas om på hela datan med den bästa hyperparameter-kombinationen.

___________________________________________________________________________________

f) Vad är kategorisk data och hur hanteras det? I ditt svar, använd begreppen nominal data, ordinal data, *one-hot-encoding*, *dummy-variable-encoding* och *ordinal encoding*. 


svar: När vi tränar ML-modeller, så behöver modellerna numerisk indata.
När vi vill använda kategorisk data såsom färg eller kön, måste vi omvandla den till numerisk data.

Nominal data är värden som saknar inbördes rangordning, (röd, grön, blå, svart). Sån här typ av data kan vi omvandla till numerisk data genom metoder som kallas för:

"one-hot-encoding" och
"dummy-variable-encoding".

Ordinal data är värden som HAR en inbördes ordning, tex {ej nöjd, neutral, nöjd}. Denna typ av data kan omvandlas till numerisk data genom en metod som kallas för ordinal-encoding.

Om datan är nominal eller ordinal beror på situation och hur vi väljer att tolka den.

När vi ska använda nominal data i en regressionsmodell så använder vi dummy-variable-encoding och för övriga modeller gäller one-hot-encoding.

svar: Principle of parismony (Occam's razor)

4. Vad menas med att "en modell är en förenkling av verkligheten"?


Om man gör en modell som mäter lycka baserat på inkomst, så är det en förenkling av verkligheten. Detta eftersom det finns flera andra faktorer som påverkar lycka, såsom relationer och hälsa tex.

Målet är inte att göra en perfekt modell utan en modell som hjälper oss att uppnå det syfte vi har.

___________________________________________________________________________________

5. Vad menas med att en modell är "överanpassad" eller *overfitted* på engelska? 


Overfitted - Modeller som är bra på träningsdatan men dåliga på valideringsdatan sägs vara överanpassade, overfitted.

Med träningsdata kan vi skapa perfekta modeller som kan genomföra alla prediktioner korrekt, vilket inte är till någon nytta om modellen inte kan prediktera ny, osedd data korrekt.

6. Högre är bättre i *scikit-learn scoring*, vad innebär det? 


scikit-learn följer den allmänna principen att "högre värden är bättre", men när vi använder MSE så är lägre poäng bättre, mindre fel. Därför använder vi negativa värden för att komma närmre 0.

100 vs 150 = 150 är bättre
-100 vs -150 = -100 är bättre (närmare 0)

7. Vad är tvärsnittsdata, tidsseriedata och paneldata? Exemplifiera när respektive datakategori kan uppstå. 


Tvärsnittsdata - data där observationer för olika individer (människor, företag, länder) samlas in vid en tidpunkt. (1 tidpunkt)

Tidsseriedata - data där observationer över tid samlas in för en individ. (över tid)

Paneldata - observationer över tid.
Tvärsnittsdata är en tidpunkt från paneldatan.
Tidsseriedatan är en individ frånb paneldatan.
