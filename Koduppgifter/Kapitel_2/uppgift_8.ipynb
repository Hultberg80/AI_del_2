````python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from joblib import dump, load
````

````python
X, y = make_regression(n_samples=20000, n_features=3, noise=0.1)
````

````python
model = LinearRegression().fit(X, y)
dump(model, "linear_model.joblib")
model_loaded = load("linear_model.joblib")

print(model_loaded.predict(X[:5]))
````

Svar:

Genom att spara modellen, så slipper vi träna den varje gång.<br>
Det sparar tid.<br>
Vi är säkra på att få samma resultat varje gång, istället för<br>
slumpmässiga resultat.
