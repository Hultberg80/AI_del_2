```python
import pandas as pd
```

```python
df = pd.read_csv("data_01.csv")
```

```python
df.head()
```

```python
# 9b

target_col = "target"
X = df.drop(columns=[target_col])
y = df[target_col]
X.shape, y.shape
```

```python
# 9c
from sklearn.model_selection import train_test_split
```

```python
# plocka ut 20% test
X_trainval, X_test, y_trainval, y_test = train_test_split(
    X, 
    y, 
    test_size=0.20, 
    random_state=42
)
```

```python
# av det som är kvar, ta 15% till VALIDATION
X_train, X_val, y_train, y_val = train_test_split(
    X_trainval,
    y_trainval,
    test_size=0.15,
    random_state=42
)

X_train.shape, X_val.shape, X_test.shape
```

```python
# 9d, Träna två regressionsmodeller

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

lin = LinearRegression()
tree = DecisionTreeRegressor(random_state=42)

lin.fit(X_train, y_train)
tree.fit(X_train, y_train)
```

```python
from sklearn.datasets import load_diabetes
from sklearn.svm import LinearSVR
from sklearn.metrics import root_mean_squared_error
```

```python
# 9e, Utvärdera på valideringsdata (RMSE)


# LinearRegression
lin_val_pred = lin.predict(X_val)
lin_rmse = root_mean_squared_error(
    y_val,
    lin_val_pred
)

# DecisionTreeRegressor
tree_val_pred = tree.predict(X_val)
tree_rmse = root_mean_squared_error(
    y_val,
    tree_val_pred
)

print("Validering - RMSE:")
print("LinearRegression:", lin_rmse)
print("DecisionTreeRegressor:", tree_rmse)
```

```python
# 9f, Träna om den bästa på train+val

import numpy as np
import pandas as pd

# slå ihop train och val
X_train_full = pd.concat([X_train, X_val], axis=0)
y_train_full = pd.concat([y_train, y_val], axis=0)

best_name = "lin" if lin_rmse <= tree_rmse else "tree"
print(f"Bästa på validering: {best_name}")

if best_name == "lin":
    best_model = LinearRegression()
else:
    best_model = DecisionTreeRegressor(random_state=42)

best_model.fit(X_train_full, y_train_full)
```
