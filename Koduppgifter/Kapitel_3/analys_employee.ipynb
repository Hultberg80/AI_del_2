#  Analys av medarbetardata

Syftet med analysen är att förstå vad som påverkar om medarbetare stannar eller lämnar företaget.  
Resultatet bygger på undersökning av nyckelvariabler som nöjdhet, arbetstimmar, antal projekt och karriärmöjligheter.  

---

##  Nyckelinsikter

- **Nöjdhet (satisfaction_level)**  
  Medarbetare som lämnat har generellt mycket lägre nöjdhetspoäng.  
  → Missnöje är en stark indikator för risk att sluta.  

- **Arbetstimmar (average_montly_hours)**  
  Hög arbetsbelastning (mycket höga timmar) syns tydligt bland de som lämnat.  
  → Extrem arbetsbelastning är en riskfaktor.  

- **Antal projekt (number_project)**  
  Medarbetare med många parallella projekt tenderar att lämna oftare.  
  → Arbetsbelastning och stress verkar bidra till avgångar.  

- **Tid i företaget (time_spend_company)**  
  Många lämnar efter ca 4–5 år.  
  → Kritisk punkt där retention-åtgärder bör sättas in.  

- **Befordran (promotion_last_5years)**  
  Få som lämnat har fått befordran.  
  → Brist på utvecklingsmöjligheter verkar driva personalomsättning.  

---

##  Visualiseringar

Nedan visas några exempel som stödjer insikterna:



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Läs in data
df = pd.read_excel("hr_employee_data.xlsx")

# 1. Satisfaction vs Left
sns.boxplot(x="left", y="satisfaction_level", data=df)
plt.title("Medarbetarnöjdhet (satisfaction_level) vs Left")
plt.show()

# 2. Average monthly hours vs Left
sns.boxplot(x="left", y="average_montly_hours", data=df)
plt.title("Arbetstimmar per månad vs Left")
plt.show()

# 3. Number of projects vs Left
sns.boxplot(x="left", y="number_project", data=df)
plt.title("Antal projekt vs Left")
plt.show()

# 4. Promotion vs Left (countplot)
sns.countplot(x="promotion_last_5years", hue="left", data=df)
plt.title("Befordran senaste 5 åren vs Left")
plt.show()
```

##  Rekommendationer till ledningsgruppen

1. **Mät och stärk medarbetarnöjdheten**  
   - Inför regelbundna engagemangsundersökningar.  
   - Sätt in åtgärder tidigt vid låg nöjdhet.  

2. **Hantera arbetsbelastning**  
   - Följ upp arbetstimmar och projektantal.  
   - Undvik att medarbetare får för många parallella projekt.  

3. **Karriärutveckling och befordran**  
   - Skapa tydliga karriärvägar och fler befordringsmöjligheter.  
   - Brist på utveckling ökar risken att medarbetare lämnar.  

4. **Retention kring 4–5 års anställning**  
   - Stärk utvecklingssamtal och karriärplanering särskilt vid denna tidpunkt.  

---

 *Sammanfattning:*  
Datan visar att personalomsättning i hög grad drivs av **nöjdhet, arbetsbelastning och karriärmöjligheter**.  
Genom att adressera dessa områden kan företaget minska risken att förlora nyckelkompetens och förbättra medarbetarretentionen.
